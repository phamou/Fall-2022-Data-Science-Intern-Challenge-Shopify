```python
# Regular exploratory data analysis (EDA) and plotting libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# read file
df = pd.read_csv("2019 Winter Data Science Intern Challenge Data Set - Sheet1.csv")
df
```

```python
# check for null values
df.isna().sum()
```

```python
# summary statistics
df.describe()
```

```python
# check for outlier values
sorted_data = df.sort_values(by=['order_amount', 'total_items'], ascending=False)
sorted_data.head(25)
```

```python
# unique prices for sneakers
sorted_data['price_per_item'] = sorted_data.order_amount/sorted_data.total_items
sorted_data.price_per_item.unique()
```

```python
# Look at orders with high total_items count
sorted_data.sort_values(by=['total_items'], ascending=False).head(25)
```

```python
# look if there are reasonable orders with sneakers priced at $352 
sorted_data.loc[sorted_data['price_per_item'] == 352]
```

```python
#add price per item in original df and eliminate outliers 
df['price_per_item'] = df.order_amount/df.total_items
df = df[df['total_items'] != 2000]
df = df[df['price_per_item'] != 25725]
df

```
